```
import numpy  as np
import pandas as pd
from sklearn import linear_model
import sklearn
import matplotlib.pyplot as plt
```

```
csvdata = pd.read_csv('kc_house_data.csv')
csvdata.keys() # check the keys available in the dataset
```

```
# drop all the features which does not plays an important role in the  model
csvdata = csvdata.drop(labels=['id','date','yr_built','yr_renovated','lat','long','zipcode', 'sqft_living15', 'sqft_lot15', 'grade',
       'sqft_above', 'sqft_basement', 'sqft_living','sqft_lot'],axis=1)
```

```
csvdata.fillna(csvdata.mean()) # Cleaning Process -> Filling NaN values with the mean values of that feature
```

```
X_train,X_test,y_train,y_test = sklearn.model_selection.train_test_split(csvdata.drop('price',axis=1),csvdata['price'],test_size=0.33)
```

```
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
regressor = linear_model.LinearRegression()
```

```
regressor.fit(X_train,y_train)
```

```
regressor.coef_
```

```
21395.41605922*3 + 212293.65269671*1 + 27979.66952114*1+ 594368.83368724*0+  122397.83993848*0+  50769.50582551 *3 
```

```
y_predicted = regressor.predict(X_test)
```

```
plt.scatter(y_test,y_predicted)
```

```
#Checking accuracy
print('Train Score : ', regressor.score(X_train,y_train) )
print('Test Score : ', regressor.score(X_test,y_test))
```
